FILTERS: 
The filtering is done by a simple sliding-window spatial filter 
that replaces the center value in the window with a new value 
(depending on the filter). The window, or kernel, is usually 
square but can be any shape.

MEAN filter: 
Takes the average (mean) of all the pixel values in the window. 

Median filter: 
Takes the median value of all the pixel values in the window. 

GAUSSIAN filter:
It represents the shape of a Gaussian bell-shaped hump. The weights are decreaseing the further away form the center of the kernel we go.


```python
import cv2
import matplotlib.pyplot as plt
from skimage import util
from scipy import ndimage
from scipy.ndimage.filters import gaussian_filter
```

```python
# Load image
image = cv2.imread('kixcodes.jpg')

# change the image to 2 dimensions -> black and white
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Add noise to the image
img_noise = util.random_noise(img_gray, mode='gaussian', seed=None, clip=True)



```

```python

# show result
fig1 = plt.figure(figsize = (15,15))

ax1 = fig1.add_subplot(121)
ax1.imshow(img_gray, interpolation = "None", cmap = "gray")
ax1.set_title("Original image")   

ax2 = fig1.add_subplot(122)
ax2.imshow(img_noise, interpolation= "None", cmap = "gray")
ax2.set_title("Salt and Pepper noise")

plt.show()
```

```python
image_median = img_noise.copy()
image_mean = img_noise.copy()
image_gaussian = img_noise.copy()
```

```python
# use the mean filter, a 3x3 matrix that takes in all the values and divides the output by 9

image_mean = cv2.blur(image_mean,(3,3))

#show the difference

fig1 = plt.figure(figsize = (15,15))

ax1 = fig1.add_subplot(121)
ax1.imshow(img_gray, interpolation = "None", cmap = "gray")
ax1.set_title("Original image") 

ax2 = fig1.add_subplot(122)
ax2.imshow(image_mean, interpolation= "None", cmap = "gray")
ax2.set_title("Mean filtered image")
```

```python
# use the meadian filter, a 3x3 matrix that takes in all the values and outputs the median value 
image_median = ndimage.median_filter(image_median, 3)


#show the difference
fig1 = plt.figure(figsize = (15,15))

ax1 = fig1.add_subplot(121)
ax1.imshow(img_gray, interpolation = "None", cmap = "gray")
ax1.set_title("Original image") 

ax2 = fig1.add_subplot(122)
ax2.imshow(image_median, interpolation= "None", cmap = "gray")
ax2.set_title("Median filtered image")
```

```python
# use a gaussian blurr 

image_gaussian = gaussian_filter(image_gaussian, 3)
#show the difference
fig1 = plt.figure(figsize = (15,15))

ax1 = fig1.add_subplot(121)
ax1.imshow(img_gray, interpolation = "None", cmap = "gray")
ax1.set_title("Original image") 

ax2 = fig1.add_subplot(122)
ax2.imshow(image_gaussian, interpolation= "None", cmap = "gray")
ax2.set_title("Gaussian filtered image")
```

```python
fig2 = plt.figure(figsize = (20,20))



ax1 = fig2.add_subplot(141)
ax1.imshow(img_gray, interpolation='none',  cmap='gray')
ax1.set_title('Original image')

ax2 = fig2.add_subplot(142)
ax2.imshow(image_median, interpolation='none',  cmap='gray')
ax2.set_title('Median filter')

ax3 = fig2.add_subplot(143)
ax3.imshow(image_mean, interpolation='none',  cmap='gray')
ax3.set_title('Mean filter')

ax4 = fig2.add_subplot(144)
ax4.imshow(image_gaussian, interpolation='none',  cmap='gray')
ax4.set_title('Gaussian filter')
plt.show()
```
