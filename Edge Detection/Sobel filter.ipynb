```python
from skimage.data import imread
import matplotlib.pyplot as plt
from scipy import ndimage
import cv2
import numpy as np
from skimage.filters import sobel
from skimage.data import camera

```

```python
# Load the image
image = imread('kixcodes.jpg')

# show original image
fig1 = plt.figure(figsize = (15,15))
ax1 = fig1.add_subplot(121)
ax1.imshow(image, interpolation = "None", cmap = "gray")
ax1.set_title("Original image")   

plt.show()
```

```python
# Make a grey scaled image - we just care about the edges
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Add the Sobel filter. Here it will calculate the derivative in the X-axis and the Y-axis, 
# compute the magnitude
# it takes some time...
edge_sobel = sobel(img_gray)

```

```python
fig1 = plt.figure(figsize = (15,15))
ax1 = fig1.add_subplot(121)
ax1.imshow(img_gray, interpolation = "None", cmap = "gray")
ax1.set_title("Original image")   

ax2 = fig1.add_subplot(122)
ax2.imshow(edge_sobel, interpolation = "None", cmap = "gray")
ax2.set_title("Sobel filter")
plt.show()

```
